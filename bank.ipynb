```python
import numpy as np 
import pandas as pd 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
bank= pd.read_csv(r'E:\myproject\Bank_Churn.csv')
bank.head(10)
```

```python
bins = [0, 20000, 50000, 100000, 150000, 200000]
labels = ['Low', 'Lower-Mid', 'Mid', 'Upper-Mid', 'High']
bank['Income_Category'] = pd.cut(bank['EstimatedSalary'], bins=bins, labels=labels)
bank.tail(10)
```

```python
bank.describe()
```

```python
bank.info()
```

```python
bank.isnull().sum()
```

```python
bank['Exited'].value_counts().plot(kind='bar', color=['red', 'green'])
```

```python
categorical_vars = ['Geography', 'HasCrCard', 'IsActiveMember', 'Income_Category', 'Gender']

for var in categorical_vars:
    plt.figure(figsize=(8, 5))
    churn_rate = bank.groupby(var)['Exited'].mean().reset_index()
    ax = sns.barplot(x=var, y='Exited', data=churn_rate, palette='viridis')
    for i, bar in enumerate(ax.patches):
        height = bar.get_height()
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            height - 0.05,  
            f'{height:.2f}',
            ha='center',
            va='top',
            color='white',
            fontsize=10,
            fontweight='bold'
        )
    plt.title(f'Churn Rate by {var}')
    plt.ylabel('Churn Rate')
    plt.xlabel(var)
    plt.ylim(0, 1)
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

```python
cot = ['Geography', 'HasCrCard', 'IsActiveMember', 'Income_Category', 'Gender','Tenure','NumOfProducts']
for vary in cot:
    plt.figure(figsize=(6, 4))
    bank[vary].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Distribution of {vary}')
    plt.xlabel(vary)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
    
```

```python
column_num= ['CreditScore', 'Age', 'Balance', 'EstimatedSalary', 'Tenure', 'NumOfProducts']
correlation_matrix = bank[column_num].corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Matrix')
```
